```python
import matplotlib.pyplot as plt
import pickle
from config import experiment_variable
```

```python

with open(f'output/GPR/{experiment_variable}_experiment_results/{experiment_variable}_experiment.pkl', 'rb') as f:
    data = pickle.load(f)
    print(data)
```

```python
# Extract learning rates and calculate average distances
rates = []
avg_distances = []

print(data.keys())

for value, experiments in data.items():
    rates.append(value)
    
    # Calculate average distance across all experiments for this learning rate
    total_distance = 0
    total_samples = 0
    
    print(experiments.keys())
    for exp_name, exp_data in experiments.items():
        total_distance += exp_data['distance_deg']
        total_samples += 1
        
        # Calculate total distance for all samples in this experiment
        #exp_distance = sum(sample_data['distance_deg'] for sample_data in exp.values())
        #total_distance += exp_distance
        #total_samples += len(exp)
    
    # Average across all samples for this learning rate
    avg_distance = total_distance / total_samples
    avg_distances.append(avg_distance)

print(f"Learning rates: {rates}")
print(f"Average distances: {avg_distances}")

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(rates, avg_distances, marker='o', linewidth=2, markersize=8)
plt.xlabel(str(experiment_variable))
plt.ylabel('Average Distance per Test Sample')
plt.title(f'Average Distance per Test Sample vs {experiment_variable}')
plt.grid(True, alpha=0.3)
plt.savefig(f'output/GPR/{experiment_variable}_experiment_results/{experiment_variable}_average_distance_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```
